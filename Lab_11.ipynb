# Travelling Salesman Problem using Dynamic Programming.
Problem Description:
Given a set of cities and distances, find the shortest possible route that visits each city once and returns to the starting city.

#Theory
TSP uses dynamic programming with bit masking. Store results of visited subsets to reduce repeated work.

## Algorithm
1. Start from source city
2. Mark visited cities using bit mask
3. Try all unvisited cities
4. Choose minimum cost path
5. Return to source city

```python
def tsp(graph):
    n = len(graph)
    dp = [[-1]*(1<<n) for _ in range(n)]

    def visit(city, mask):
        if mask == (1<<n) - 1:
            return graph[city][0]

        if dp[city][mask] != -1:
            return dp[city][mask]

        ans = float('inf')

        for next_city in range(n):
            if mask & (1<<next_city) == 0:
                new_cost = graph[city][next_city] + visit(next_city, mask | (1<<next_city))
                if new_cost < ans:
                    ans = new_cost

        dp[city][mask] = ans
        return ans

    return visit(0, 1)
```

```python
graph = [
[0, 10, 15, 20],
[10, 0, 35, 25],
[15, 35, 0, 30],
[20, 25, 30, 0]
]

print("Minimum travelling cost:", tsp(graph))
```

# Exercise
1. Print path of cities visited
2. Take graph input from user
3. Test with different number of cities

```python
def tsp_with_path():
    try:
        n = int(input("Enter number of cities: "))
        print("Enter the adjacency matrix (row by row, space separated):")
        graph = []
        for i in range(n):
            graph.append(list(map(int, input().split())))
    except ValueError:
        print("Invalid input. Please enter numbers only.")
        return

    # dp stores the minimum cost
    dp = [[-1] * (1 << n) for _ in range(n)]
    # parent stores the next_city chosen to get that minimum cost
    parent = [[-1] * (1 << n) for _ in range(n)]

    def visit(city, mask):
        # Base Case: All cities visited, return to start (0)
        if mask == (1 << n) - 1:
            return graph[city][0]

        if dp[city][mask] != -1:
            return dp[city][mask]

        ans = float('inf')
        best_next_city = -1

        for next_city in range(n):
            # Check if next_city is unvisited
            if (mask & (1 << next_city)) == 0:
                new_cost = graph[city][next_city] + visit(next_city, mask | (1 << next_city))

                if new_cost < ans:
                    ans = new_cost
                    best_next_city = next_city

        dp[city][mask] = ans
        parent[city][mask] = best_next_city # Best decision store kar rahe hain
        return ans

    # 1. Calculate Min Cost
    min_cost = visit(0, 1)

    # 2. Reconstruct Path using parent table
    path = []
    curr_city = 0
    curr_mask = 1

    while curr_city != -1:
        path.append(curr_city)
        curr_city = parent[curr_city][curr_mask]
        if curr_city != -1:
            curr_mask |= (1 << curr_city)

    path.append(0) # Return back to start

    print(f"\nMinimum travelling cost: {min_cost}")
    print(f"Optimal Path: {' -> '.join(map(str, path))}")

# Run the exercise
tsp_with_path()
```
